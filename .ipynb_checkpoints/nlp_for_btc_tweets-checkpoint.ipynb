Define the clean text function for, well, clean text

```python
stop_words = nltk.corpus.stopwords.words(['english'])
lem = WordNetLemmatizer()

def cleaning(data):
    '''
    Clean tweet text data
    Arg:
    data: raw tweet text data
    Return:
    data: cleaned tweet text data
    '''
    # make sure data is a string
    data = str(data)
    
    #remove urls
    tweet_without_url = re.sub(r'http\S+',' ', data)

    #remove hashtags
    tweet_without_hashtag = re.sub(r'#\w+', ' ', tweet_without_url)

    #3. Remove mentions and characters that not in the English alphabets
    tweet_without_mentions = re.sub(r'@\w+',' ', tweet_without_hashtag)
    precleaned_tweet = re.sub('[^A-Za-z]+', ' ', tweet_without_mentions)

    #2. Tokenize
    tweet_tokens = TweetTokenizer().tokenize(precleaned_tweet)

    #3. Remove Puncs
    tokens_without_punc = [w for w in tweet_tokens if w.isalpha()]

    #4. lemma
    tokens_without_lemma = [lem.lemmatize(t) for t in tokens_without_punc]

    #5. Removing Stopwords
    tokens_without_stopwords = [t for t in tokens_without_lemma if t not in stop_words]

    #6. Lower word
    text_cleaned = [t.lower() for t in tokens_without_stopwords]
    
    #6. Joining
    return " ".join(text_cleaned)
```

In this section, we gonna calculate some of the index to make a baseline for further analysis. This include the **compound index** (utilize the VADER mode) and the **score index** (base on the compound index with the inclusion of all the other factor in the dataset)

### Compound index

Calculate the compound index and add to a new column of df

```python
# Calculate the compound index
analyzer = SentimentIntensityAnalyzer()
compound = []
for i,s in enumerate(tqdm(df['tweets'],position=0, leave=True)):
    # print(i,s)
    vs = analyzer.polarity_scores(str(s))
    compound.append(vs["compound"])
df["compound"] = compound
```

> Successfully calculated compound index

### Score index

The author propose a new index which include the calculation of all the other factor in the dataset, which fomular is mentioned in the paper. And because we already have all the variables to calculate it, we shall now do it for the convenient of resample the df and visualize in the later steps

```python
scores = []
for i, s in tqdm(df.iterrows(), total=df.shape[0],position=0, leave=True):
    try:
        score = ((s["compound"] * ( ((int(s["user_favourites"]))*0.8) + ((int(s["user_favourites"]))*10*0.2) + (int(s["is_retweet"])+1) + (1+(int(s["user_followers"])*0.001)) ) * (1.3 if ((int(s["user_verified"])) == 1) else 1)))
        scores.append(score)
    except:
        scores.append(np.nan)
df["score"] = scores
```

Now we create 2 functions to apply for the bitcoin price movement. 
- The first one crypto_price_cate is used for classify the trend of the crypto price base on the second function.
- The 2nd functino is used for calculate the difference of the price base on the given interval time.
- The observe_period parameter is used to set the interval time period of the 2nd function. (Ex: 1 == 1 days interval of the price)

```python
def crypto_price_cate(score):
    """
    Categorize the price movement of a cryptocurrency based on a score.

    Parameters:
    score (float): The score representing the price movement. 
                   - If the score is less than 1, the movement is considered 'negative'.
                   - If the score is exactly 1, the movement is considered 'neutral'.
                   - If the score is greater than 1, the movement is considered 'positive'.

    Returns:
    str: The category of the price movement ('negative', 'neutral', or 'positive').
    """
    if score < 1:
        return 'negative'
    elif score == 1:
        return 'neutral'
    else:
        return 'positive'
def observe_period(period):
    """
    Observe and categorize the cryptocurrency price movement over a given period.

    Parameters:
    period (int): The period over which to observe the price movement.

    Returns:
    pandas.Series: A Series with the categorized price movements over the specified period.
    """
    res = crypto_usd_grouped.shift(period)/crypto_usd_grouped
    res = res.apply(crypto_price_cate)
    return res 

# Add a new date_clean column to df because the date column have time in it
df['date_clean'] = pd.to_datetime(df['date'], errors='coerce').dt.strftime('%Y-%m-%d')
df['date_clean'] = pd.to_datetime(df['date_clean'])

# Compare price ratio in 1 days. price_1_days_later/ price_now 
time_sentiment = observe_period(1) 
df['crypto_trend'] = df.date_clean.apply(lambda x: time_sentiment[x] if x in time_sentiment else np.nan)
```

> Successfully defined fuctions

Define the function for determine the sentiment base on the given score, here is the compound score. As for the documentation of the VADER model, the classification should apply like explain in the parameter segment of the docstring (source: Author of VADER model)

```python
def getSentiment(score):
    """
    Determine the sentiment based on the given score.

    Parameters:
    score (int or float): The score representing sentiment.
                          - If the score is less than 0, the sentiment is 'negative'.
                          - If the score is 0, the sentiment is 'neutral'.
                          - If the score is greater than 0, the sentiment is 'positive'.

    Returns:
    str: The sentiment ('negative', 'neutral', or 'positive') based on the score.
    """
    if score < 0:
        return 'negative'
    elif score == 0:
        return 'neutral'
    else:
        return 'positive'

# Create a new column named 'sentiment' for label
df['sentiment'] = df['compound'].apply(getSentiment)
```

### Examine Sentiment score and price correlation

However, the compound score and sentiment label only evaluate the sentiment of each individual tweet without considering the level of influence and impact of that tweet on other users. Therefore, we will utilize a new integrated metric that incorporates both factors: tweet sentiment and level of influence. To calculate the level of influence of a tweet, various factors come into play, such as the credibility of the poster, the number of followers, the level of engagement of the tweet, the number of views, shares, etc. After considering the factors affecting the level of influence of a tweet, the author proposes to use a formula referenced from Hasan et al., 2024, and simplified to compute the new metric below, termed as the "score". The score number is already calculated above, so in the next part we will visualize it in the graph to observe the different between compound index and score index, and then run 3 of the cross-correlation methods to evaluate the correlation effect of bitcoin price and twitter sentiments.

```python
# Normalize time series data
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
score_scaled = min_max_scaler.fit_transform(tweets_grouped['score'].values.reshape(-1,1))
tweets_grouped['normalized_score'] = score_scaled
crypto_used_grouped_scaled = min_max_scaler.fit_transform(crypto_usd_grouped.values.reshape(-1,1))

# plot graph
fig, ax1 = plt.subplots(figsize=(20,10))
ax1.set_title("Sentiment score and Bitcoin price", fontsize=18)
ax1.tick_params(labelsize=14)

ax2 = ax1.twinx()
ax1.plot_date(tweets_grouped.index, tweets_grouped['normalized_score'], 'g-')
ax2.plot_date(crypto_usd_grouped.index, crypto_used_grouped_scaled, 'b-')

ax1.set_ylabel("Sentiment", color='g', fontsize=16)
ax1.set_xlabel("Time period", fontsize=16)
ax2.set_ylabel("Bitcoin normalized", color='b', fontsize=16)
plt.show()
```

> As we can see, the “score” coefficient has displayed negative values within the normalized range from -1 to 1. However, due to the predominance of positive tweets, the community consistently expresses market optimism.

### Examine correlation using mutiple cross-correlation methods

Now we run the 3 method of cross-correlation: Pearson, Kendall and Spearman to determine the correlation between bitcoin price and twitter sentiment

```python
import matplotlib.pyplot as plt

# Assuming 'score' column contains sentiment scores
sentiment_scores = tweets_grouped['score']

# Assuming 'crypto_price' column contains cryptocurrency prices
crypto_prices = crypto_usd_grouped

# Define the cross-correlation function
def crosscorr(datax, datay, lag=0, method="pearson"):
    return datax.corr(datay.shift(lag), method=method)

# Calculate cross-correlation using different methods
methods = ["pearson", "kendall", "spearman"]
for method in methods:
    xcov = [crosscorr(sentiment_scores, crypto_prices, lag=i, method=method) for i in range(-20, 20)]
    plt.plot(range(-20, 20), xcov)
    plt.title(f"{method.capitalize()} cross-correlation")
    plt.xlabel("lag")
    plt.ylabel("correlation")
    plt.show()
```

```python
# Derivative
tweets_grouped_derivative = pd.Series(np.gradient(tweets_grouped['normalized_score'].values), tweets_grouped.index, name='slope')
crypto_usd_grouped_derivative = pd.Series(np.gradient(crypto_usd_grouped.values), crypto_usd_grouped.index, name='slope')
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1.set_title("Sentiment score and Bitcoin price fluctuation", fontsize=18)
ax1.set_xlabel("Time period", fontsize=16)
ax1.tick_params(labelsize=14)

ax2 = ax1.twinx()
ax1.plot_date(tweets_grouped_derivative.index, tweets_grouped_derivative, 'g-')
ax2.plot_date(crypto_usd_grouped_derivative.index, crypto_usd_grouped_derivative, 'b-')

ax1.set_ylabel("Sentiment's derivative", color='g', fontsize=16)
ax2.set_ylabel('Bitcoin price derivative', color='b', fontsize=16)
plt.show()
plt.('derivative_crypto_currency_vs_twitter_sentiment', dpi=1000)
```

```python
def crosscorr(datax, datay, lag=0, method="pearson"):
    return datax.corr(datay.shift(lag), method=method)

# Calculate cross-correlation using different methods
methods = ["pearson", "kendall", "spearman"]
for method in methods:
    xcov = [crosscorr(tweets_grouped['score'], crypto_usd_grouped_derivative, lag=i, method=method) for i in range(-20, 20)]
    plt.plot(range(-20, 20), xcov)
    plt.title(f"{method.capitalize()} cross-correlation (derivative)")
    plt.xlabel("lag")
    plt.ylabel("correlation")
    plt.show()
```

Create the new 2 columns for value aroon up and down as new variables. The fomular is mentioned in the paper. 

```python
# aroon for sample
from sklearn.preprocessing import MinMaxScaler

# Create a copy of the DataFrame 'data'
df_aroon_test = data.copy()

# Calculate Aroon values for each date in df_aroon_test
aroon_period = 30  # Aroon period
aroon_up = []
aroon_down = []

for date in df_aroon_test.index:
    # Get the corresponding Bitcoin price data for the date within the Aroon period
    btc_data_period = df_aroon_test.loc[date - pd.DateOffset(days=aroon_period - 1):date]
    
    # Calculate highest high and lowest low within the Aroon period
    highest_high = btc_data_period['High'].max()
    lowest_low = btc_data_period['Low'].min()
    
    # Calculate number of days since highest high and lowest low
    days_since_high = (btc_data_period['High'] == highest_high).idxmax() - date
    days_since_low = (btc_data_period['Low'] == lowest_low).idxmax() - date
    
    aroon_up_value = ((aroon_period - days_since_high.days) / aroon_period) * 100
    aroon_down_value = ((aroon_period - days_since_low.days) / aroon_period) * 100
    
    aroon_up.append(aroon_up_value)
    aroon_down.append(aroon_down_value)

# Normalize Aroon values
scaler = MinMaxScaler()
aroon_up_normalized = scaler.fit_transform(pd.DataFrame(aroon_up))
aroon_down_normalized = scaler.fit_transform(pd.DataFrame(aroon_down))

# Append normalized Aroon values to df_aroon_test
df_aroon_test['aroon_up'] = aroon_up_normalized
df_aroon_test['aroon_down'] = aroon_down_normalized

# Print df_aroon_test with appended Aroon values
df_aroon_test.head(10)
```

## 3. Calculate ADL value

Create the new 1 columns for value ADL as new variables. The fomular is mentioned in the paper.

```python
# Copy a new dataframe for calculation
agg_data = data.copy()

# Aggregate data by month
agg_data['Month_Year'] = agg_data.index.to_period('M')  # Add a column for month and year
agg_data_monthly = agg_data.groupby('Month_Year').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
})

# Calculate Money Flow Multiplier (MF Multiplier)
mf_multiplier = ((agg_data_monthly['Close'] - agg_data_monthly['Low']) - (agg_data_monthly['High'] - agg_data_monthly['Close'])) / (agg_data_monthly['High'] - agg_data_monthly['Low'])

# Calculate Money Flow Volume (MF Volume)
mf_volume = mf_multiplier * agg_data_monthly['Volume']

# Calculate Accumulation/Distribution Line (ADL)
adl = mf_volume.cumsum()

# Step 2: Iterate through records in 'df' DataFrame and add ADL values
adl_values = []

# Iterate through 'df' DataFrame
for date_clean in df['date_clean']:
    # Extract month and year
    month_year = date_clean.to_period('M')
    
    # Find corresponding ADL value for the month
    if month_year in adl.index:
        adl_value = adl.loc[month_year]
        adl_values.append(adl_value)
    else:
        adl_values.append(None)  # No corresponding ADL value found

# Append ADL values to 'df' DataFrame as a new column
df['adl'] = adl_values
df.head(2)
```

## VADER Model - Calculate VADER variables

While the VADER compound score provides a single measure of sentiment, it can be valuable to delve deeper and analyze the distribution of positive, neutral, and negative sentiment within a text. This allows for a more nuanced understanding of the emotional landscape expressed in tweets.  


In this part, we decompose the VADER compound score into three separate variables: positive score, neutral score, and negative score. These scores represent the relative proportions of positive, neutral, and negative sentiment within a tweet. This division will allow us to analyze tweets from multiple dimensions rather than just one dimension of the compound index. A tweet can have from 1 to 3 indices with different weights. The compound index is removed due to multicollinearity problem concern, but the score index is still utilized as a variable, albeit calculated from the compound angle coefficient, it is combined with other data.


```python
analyzer = SentimentIntensityAnalyzer()

def analyze_sentiment_pos(tweet):
    """
    Analyze the positive sentiment of a tweet.

    Parameters:
    tweet (str): The tweet text.

    Returns:
    str: The positive sentiment score as a string.
    """
    sentiment_dict = analyzer.polarity_scores(tweet)
    return str(sentiment_dict['pos'])

def analyze_sentiment_neg(tweet):
    """
    Analyze the negative sentiment of a tweet.

    Parameters:
    tweet (str): The tweet text.

    Returns:
    float: The negative sentiment score.
    """
    sentiment_dict = analyzer.polarity_scores(tweet)
    return sentiment_dict['neg']

def analyze_sentiment_neu(tweet):
    """
    Analyze the neutral sentiment of a tweet.

    Parameters:
    tweet (str): The tweet text.

    Returns:
    float: The neutral sentiment score.
    """
    sentiment_dict = analyzer.polarity_scores(tweet)
    return sentiment_dict['neu']

def analyze_sentiment_comp(tweet):
    """
    Analyze the compound sentiment of a tweet.

    Parameters:
    tweet (str): The tweet text.

    Returns:
    float: The compound sentiment score.
    """
    sentiment_dict = analyzer.polarity_scores(tweet)
    return sentiment_dict['compound']
```

Calculate the comp, pos, neg and neu value using VADER model function

```python
df2 = pd.DataFrame()
df2["Vader_Comp"] = df['cleaned_tweets'].apply(analyze_sentiment_comp)
df2["Vader_Pos"] = df['cleaned_tweets'].apply(analyze_sentiment_pos)
df2["Vader_Neg"] = df['cleaned_tweets'].apply(analyze_sentiment_neg)
df2["Vader_Neu"] =df['cleaned_tweets'].apply(analyze_sentiment_neu)
df2.shape
df = pd.concat([df, df2], axis=1)
df.head(2)
```

# 6. Conclusion

> - (i) There exists a correlation between user sentiment and Bitcoin price movement; however, the impact on the trend of the two variables on each other is still not entirely clear.
> - (ii) The LSTM deep learning model has the ability to classify sentiment down to the level of individual words for tweet content effectively, demonstrating the model's quality in text processing. However, when training and predicting for price trend classification of Bitcoin, the model did not identify many common patterns, resulting in an accuracy level (53.7%) just slightly above random guessing. Once again, this indicates that there is no added advantage in using these NLP models to classify tweets without additional support from input data, consistent with the content of the studies by (Rezaie et al., 2020), (Sattarov et al., 2020), (Kraaijeveld & De Smedt, 2020), (McNally et al., 2018) and (Edgari et al., 2022).

The ensemble machine learning models AdaBoost and XGBoost exhibited the highest accuracy levels (73.1% and 76.1%) among all models. Several other studies have also yielded optimal results with the XGBoost model, such as Sekhar et al., 2022, Edgari et al., 2022.


# 7. Limitation and further research

> While the research provides valuable insights into the relationship between tweet sentiment and Bitcoin price trends, several limitations exist that warrant further investigation and refinement in future research endeavors:
> - Text-based analysis only: Our analysis primarily focuses on textual data extracted from tweets, neglecting the potential impact of multimedia content such as images, videos, and audio. Incorporating multimedia analysis techniques could provide a more comprehensive understanding of sentiment expressed on social media platforms like Twitter.
> - Language limitations: Our study currently analyzes sentiment in English-language tweets, overlooking sentiment expressed in other languages. Expanding the analysis to include multilingual sentiment analysis may offer a more inclusive perspective, especially considering the global nature of Bitcoin trading and social media usage.
> - Limited scope: While we concentrate on Twitter data, other sources of information such as news outlets, closed-group communities (e.g., pump-and-dump groups in the cryptocurrency space), and alternative social media platforms remain untapped. Future research could explore the sentiment expressed in these additional outlets to gain a more holistic view of market sentiment.
> - Time lag considerations: Our study adopts a fixed time lag of one day to analyze the impact of tweet sentiment on Bitcoin prices. However, the time lag between sentiment expression on social media and its influence on market prices may vary and could be influenced by factors such as the virality of tweets and market reaction times. Investigating different time lag intervals could provide insights into the dynamics of sentiment-driven market movements.
> - Narrow scope of influencers: Our analysis focuses on general sentiment expressed by Twitter users and does not specifically address the influence of key opinion leaders (KOLs), public figures, politicians, or tech CEOs. Future research could delve deeper into the impact of influential individuals and entities on market sentiment and price trends within the cryptocurrency space.
> - In light of these limitations, future research endeavors could address these gaps to enhance the robustness and applicability of sentiment analysis in predicting Bitcoin price trends. By broadening the scope of analysis to encompass multimedia content, multilingual sentiment, alternative information sources, varied time lags, and specific influencer effects, researchers can develop more nuanced models for understanding and forecasting cryptocurrency market dynamics.
